```python
orders_ds = pd.read_csv(r'C:\Users\kishore\orders.csv')
orders_ds
```

```python
orders_ds.fillna({'ship_mode':0},inplace=True)
```

```python
orders_ds.isnull().sum()
```

```python
orders_ds = orders_ds.rename(columns={
    'Order Id': 'order_id',
    'Order Date':'order_date',
    'Ship Mode':'ship_mode',
    'Postal Code':'postal_code',
    'Sub Category' :'sub_category',
    'Product Id':'product_id',
    'cost price': 'cost_price',
    'List Price': 'list_price',
    'Discount Percent': 'discount_percent'
})
orders_ds.columns = orders_ds.columns.str.lower()
orders_ds
    
    
    
    
    
    
    
    
```

```python
for col in orders_ds.select_dtypes(include='object').columns:
    orders_ds[col] = orders_ds[col].map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
orders_ds['discount']=orders_ds['list_price']*orders_ds['discount_percent']*.01
orders_ds['sale_price']=orders_ds['list_price']-orders_ds['discount']
orders_ds['profit']=orders_ds['sale_price']-orders_ds['cost_price']
orders_ds.to_csv('orders_updated.csv', index=False)
orders_ds
```
